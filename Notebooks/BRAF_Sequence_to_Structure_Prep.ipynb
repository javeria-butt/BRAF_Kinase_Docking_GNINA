### **2. Automated DNA to mRNA Transcription Pipeline**

This section implements a custom structural function that reads the human BRAF genomic sequence (braf_dna.fasta), executes an in silico molecular transcription (substituting Thymine with Uracil), updates structural metadata headers to maintain integrity, and exports a pristine transcript.

```python

def transcribe_dna_to_mrna(input_fasta, output_fasta):
    """
    Reads the DNA FASTA file of BRAF, transcribes the sequence to mRNA
    (replacing Thymine 'T' with Uracil 'U'), and writes the result to a new file.
    """
    # 1. Read the DNA sequence using Biopython's SeqIO
    print(f"Reading DNA sequence from: {input_fasta}...")
    dna_record = SeqIO.read(input_fasta, "fasta")

    # 2. Perform the functional transcription step
    # Biopython's .transcribe() automatically handles the T -> U conversion
    mrna_sequence = dna_record.seq.transcribe()

    # 3. Create a new SeqRecord for the mRNA file to preserve FASTA format metadata
    mrna_record = dna_record
    mrna_record.seq = mrna_sequence
    mrna_record.id = dna_record.id + "_mRNA"
    mrna_record.description = dna_record.description + " (Transcribed to mRNA)"

    # 4. Write the final product out to a new file
    print(f"Writing transcribed mRNA sequence to: {output_fasta}...")
    with open(output_fasta, "w") as output_handle:
        SeqIO.write(mrna_record, output_handle, "fasta")

    print("Transcription successfully completed!")

# Define file pathways for the execution
input_file = "braf_dna.fasta"
output_file = "braf_mrna.fasta"

# Execute the transcription function
transcribe_dna_to_mrna(input_file, output_file)
```
